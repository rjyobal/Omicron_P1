## Poll Analysis
### Findings:
- We can perceive significant changes in public preferences over the months preceding the election.
- When we compare the polls with the election result, we realize that these are not a determinant, due to the American election system
- 2020 shows the biggest separation between candidates (almost 10 points) in the last 20 years (Presidential Elections since 2000). 

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#Reading the csv files
y2000_poll_df = pd.read_csv("Resources/2000_us_elections_polls.csv")
y2004_poll_df = pd.read_csv("Resources/2004_us_elections_polls.csv")
y2008_poll_df = pd.read_csv("Resources/2008_us_elections_polls.csv")
y2012_poll_df = pd.read_csv("Resources/2012_us_elections_polls.csv")
y2016_poll_df = pd.read_csv("Resources/2016_us_elections_polls.csv")
y2020_poll_df = pd.read_csv("Resources/2020_us_elections_polls.csv")
```

```python
#Show first Dataframe
y2000_poll_df
```

```python
#Rename Columns
y2000_poll_df.columns=['Month', 'Republicans', 'Democrats']
y2004_poll_df.columns=['Month', 'Republicans', 'Democrats']
y2008_poll_df.columns=['Month', 'Republicans', 'Democrats']
y2012_poll_df.columns=['Month', 'Republicans', 'Democrats']
y2016_poll_df.columns=['Month', 'Republicans', 'Democrats']
y2020_poll_df.columns=['Month', 'Republicans', 'Democrats']
```

```python
#Plot Year 2000
y00=y2000_poll_df.plot(color=['red','blue'])
y00.set_title("2000 US Presidential Elections Polls")
y00.set_xlabel("Months to Election Date")
y00.set_ylabel("Percentage")
y00.figure.savefig("Images/2000polls.png")
plt.show()
```
